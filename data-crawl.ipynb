```python
from google_play_scraper import Sort, reviews_all, reviews
from datetime import datetime
import pandas as pd
```

```python

result, continuation_token = reviews("kr.epopsoft.word", lang="ko", country="kr")
cnt = 0

while True:
    cnt += 1
    print(f"{cnt}번째 크롤링 중...")
    tlst, continuation_token = reviews("kr.epopsoft.word", lang="ko", country="kr", continuation_token=continuation_token)
    result += tlst
    last = result[-1]["at"]
    
    if last < datetime(2025, 1, 9):
        break
```

```python
print(f"{len(result)}개가 크롤링됐어요")
print(result[-1]["at"])
print(type(result[-1]["at"]))
```
